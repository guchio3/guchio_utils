```python
import pandas as pd
import numpy as np

import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import sys
sys.path.append("../")

import my_lightgbm as lgb
```

```python
df = sns.load_dataset("iris")
df
```

```python
le = LabelEncoder()
le.fit(df['species'].values)

train = lgb.Dataset(
    data=df.drop("species", axis=1).values,
    label=le.transform(df['species'].values),
)
```

```python
PARAMS = {
    'objective': 'multiclass',
#        'metric': 'None',
    'num_class': 3,
}
```

```python
hist, models = lgb.cv(
    train_set=train, 
    params=PARAMS,
    nfold=3,
    verbose_eval=10,
)
```

```python
np.argmax(models.boosters[0].predict(df.drop("species", axis=1).values), axis=1)
```
